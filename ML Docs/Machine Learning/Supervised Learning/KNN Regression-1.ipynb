```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"F:\50_Startups.csv")
```

### Handling missing value

```python
dataset.mean()
```

```python
dataset.fillna(dataset.mean(),inplace=True)
```

```python
dataset.isnull().any()
```

```python
dataset["State"].mode()[0]
```

```python
dataset["State"].fillna(dataset["State"].mode()[0],inplace=True)
```

```python
dataset.head()
```

```python
dataset.isnull().any()
```

```python
dataset
```

### Encoding Categorical text data

##### One Hot Encoding

```python
dataset=pd.get_dummies(dataset,columns=["State"])
```

```python
dataset
```

```python
dataset
```

#### Split Dependent and Independent Variables

```python
dataset
```

```python
#Independent Variable Input
x=dataset.drop("Profit",axis=1)
```

```python
x.head(3)
```

```python
#Dependent Variable
y=dataset["Profit"]
```

```python
y
```

### Split the data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
x_train
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#Init MinMax Scaler
sc=MinMaxScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
x_train
```

### Building the KNN Model

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model=KNeighborsRegressor(n_neighbors=3,p=2)
```

```python
knn_model.fit(x_train,y_train)
```

#### Test the model

```python
y_pred=knn_model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
y_pred
```

### MSE

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

### RMSE

```python
rmse=np.sqrt(mse)
```

```python
rmse
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
y_pred_train=knn_model.predict(x_train)
```

```python
mse=mean_squared_error(y_train,y_pred_train)
```

```python
rmse=np.sqrt(mse)
```

```python
rmse
```

```python
r2_score(y_train,y_pred_train)
```
