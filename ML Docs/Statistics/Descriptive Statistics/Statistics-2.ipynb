# Find Outliers

```python
q1=dataset["Profit"].describe()["25%"]
```

```python
q3=dataset["Profit"].describe()["75%"]
```

```python
iqr=q3-q1
```

```python
iqr
```

```python
lower_boundary=q1-(1.5*iqr)
```

```python
upper_boundary=q3+(1.5*iqr)
```

```python
lower_boundary
```

```python
upper_boundary
```

# Filter

```python
dataset[dataset["Profit"] < lower_boundary]
```

```python
dataset[dataset["Profit"] > upper_boundary]
```

```python
dataset[(dataset["Profit"] < lower_boundary) | (dataset["Profit"] > upper_boundary) ]
```
