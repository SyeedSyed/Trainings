# Step-1 Sampling

```python
import random
```

```python
random.choices(dataset["Purchase"],k=20)
```

```python
#no of samples
n=50
#sample size
k=60
```

```python
samples=[]
for i in range(0,n):
    samples.append(random.choices(dataset["Purchase"],k=k))
```

```python
samples
```

```python
sample_means=[]
```

```python
for sample in samples:
    sample_means.append(np.mean(sample))
```

```python
sample_means
```

# Confidence Interval

```python
s_mean=np.mean(sample_means)
```

```python
sns.distplot(sample_means)
```

```python
s_std=np.std(sample_means)
```

# Confidence--90%

```python
z=1.64
```

```python
lower_limit=s_mean-(z*s_std)
```

```python
lower_limit
```

```python
upper_limit=s_mean+(z*s_std)
```

```python
upper_limit
```

```python
lower_limit,upper_limit
```

```python
#We are 90% confident that the mean of the population will be in interval range of 8188 to 10263
```
