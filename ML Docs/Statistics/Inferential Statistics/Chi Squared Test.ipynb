# Chi Square Test

```python
sample_data=dataset.sample(n=10000)
```

```python
sample_data
```

```python
pd.crosstab(sample_data["Gender"],sample_data["City"],margins=True,margins_name="Total")
```

```python
observed_values=pd.crosstab(sample_data["Gender"],sample_data["City"])
```

```python
observed_values
```

```python
#converts dataframe into numpy array
observed_values=observed_values.values
```

```python
observed_values
```

```python
#1 Calculated chi square value
#2 Calculated p value
#3 DOF--Degree of Freedom
#4--Expected Values
chi_cal,p,dof,expected_values=s.chi2_contingency(observed_values)
```

```python
chi_cal
```

```python
p
```

```python
dof
```

```python
expected_values
```

```python
#Confidence 95%
#Significance Level -- 0.05
```

```python
chi_cal
```

```python
#Significance Level -- 0.05
#dof=2
chi_tab=5.99
```

```python
if(chi_cal>chi_tab):
    print("Reject the Null Hypothesis")
    print("There is a relationship between gender and city")
else:
    print("Accept the Null Hypothesis")
    print("There is no relationship between gender and city")
```

### We significantly prove that gender has no relationship with city
